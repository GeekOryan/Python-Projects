```python
# Personal financial Tracker

import pandas as pd

# Read your banking data
file_path = ""
df = pd.read_csv(file_path, skiprows=4, header=None, encoding='latin-1')
df.columns = ['Date', 'Description', 'Amount', 'Balance']

print("=== MONEY RECEIVED - FULL BREAKDOWN ===")

# Get ALL money received (all positive amounts)
all_received = df[df['Amount'] > 0]
total_received = all_received['Amount'].sum()

print(f"TOTAL MONEY RECEIVED: R{total_received:.2f}")
print(f"Number of incoming transactions: {len(all_received)}")

print("\n" + "="*60)

# Break down by categories
print("\n=== BREAKDOWN BY SOURCE ===")

# SIK money
sik_received = all_received[all_received['Description'].str.contains('SIK', case=False, na=False)]
sik_total = sik_received['Amount'].sum()
print(f"SIK: R{sik_total:.2f}")

# Deposits
deposit_received = all_received[all_received['Description'].str.contains('deposit|dep', case=False, na=False)]
deposit_total = deposit_received['Amount'].sum()
print(f"Deposits: R{deposit_total:.2f}")

# Other sources (everything else)
other_received = all_received[~all_received['Description'].str.contains('SIK|deposit|dep', case=False, na=False)]
other_total = other_received['Amount'].sum()
print(f"Other sources: R{other_total:.2f}")

print(f"\nCHECK: SIK + Deposits + Other = R{sik_total + deposit_total + other_total:.2f}")

print("\n" + "="*60)
print("=== ALL MONEY RECEIVED - SHOWING EVERY TRANSACTION ===")

# Show ALL incoming transactions sorted by amount (largest first)
all_received_sorted = all_received.sort_values('Amount', ascending=False)

for _, transaction in all_received_sorted.iterrows():
    print(f"R{transaction['Amount']:8.2f} | {transaction['Date']} | {transaction['Description']}")

print("\n" + "="*60)
print("=== LARGEST TRANSACTIONS ===")

# Show top 10 largest transactions
top_10 = all_received_sorted.head(10)
for i, (_, transaction) in enumerate(top_10.iterrows(), 1):
    print(f"{i:2d}. R{transaction['Amount']:8.2f} | {transaction['Date']} | {transaction['Description']}")

print("\n" + "="*60)
print("=== SUMMARY ===")
print(f"Total from SIK: R{sik_total:.2f}")
print(f"Total from deposits: R{deposit_total:.2f}") 
print(f"Total from other sources: R{other_total:.2f}")
print(f"GRAND TOTAL: R{total_received:.2f}")
```
