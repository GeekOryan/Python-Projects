```python
# Simple Inventory System

inventory = []
product_id_counter = 1  # The Unique identification for each product

# Show the menu
def show_menu():
    print("WELCOME TO THE INVENTORY SYSTEM")
    print("1. Add an Item to the inventory")
    print("2. View an Inventory")
    print("3. Search an Item in the inventory")
    print("4. Update an Item in the inventory")
    print("5. Delete an Item in the inventory")
    print("6. Sort the Items by Price")
    print("7. Filter the Items by Category")
    print("8. Inventory Summary Report")
    print("9. Category Summary Report")
    print("10. Exit")

# Add a new item to inventory
def add_item():
    global product_id_counter
    name = input("Enter the item name: ")
    category = input("Enter the item category: ")
    quantity = int(input("Enter the quantity: "))
    price = float(input("Enter the price per item: "))

    item = {
        "id": product_id_counter,
        "name": name,
        "category": category,
        "quantity": quantity,
        "price": price
    }
    inventory.append(item)
    product_id_counter += 1
    print("Item has been added successfully.")

# View all inventory items
def view_inventory():
    if not inventory:
        print("Inventory is empty.")
        return

    print("CURRENT INVENTORY")
    for item in inventory:
        print(f"ID: {item['id']}, Name: {item['name']}, Category: {item['category']}, Qty: {item['quantity']}, Price: R{item['price']}")

# Search for an item by name
def search_item():
    search_name = input("Enter the item name to search: ").lower()
    found = False
    for item in inventory:
        if item["name"].lower() == search_name:
            print(f"Found: ID: {item['id']}, Name: {item['name']}, Category: {item['category']}, Qty: {item['quantity']}, Price: R{item['price']}")
            found = True
            break
    if not found:
        print("Item not found in inventory.")

# Update an item
def update_item():
    try:
        item_id = int(input("Enter the item ID to update: "))
        for item in inventory:
            if item["id"] == item_id:
                item["name"] = input("Enter new name: ")
                item["category"] = input("Enter new category: ")
                item["quantity"] = int(input("Enter new quantity: "))
                item["price"] = float(input("Enter new price: "))
                print("Item updated successfully.")
                return
        print("Item not found.")
    except ValueError:
        print("Invalid input. Please enter numbers for ID, quantity, and price.")

# Delete an item
def delete_item():
    try:
        item_id = int(input("Enter the item ID to delete: "))
        for item in inventory:
            if item["id"] == item_id:
                inventory.remove(item)
                print("Item deleted successfully.")
                return
        print("Item not found.")
    except ValueError:
        print("Invalid input. Please enter a number for ID.")

# Sort items by price
def sort_by_price():
    if not inventory:
        print("The inventory is empty.")
        return
    sorted_items = sorted(inventory, key=lambda x: x["price"])
    print("Products Sorted by Price")
    for item in sorted_items:
        print(f"ID: {item['id']}, Name: {item['name']}, Price: R{item['price']}")

# Filter items by category
def filter_by_category():
    category = input("Enter category to filter: ").lower()
    filtered_items = [item for item in inventory if item["category"].lower() == category]
    if filtered_items:
        print(f"\n--- Items in Category '{category}' ---")
        for item in filtered_items:
            print(f"ID: {item['id']}, Name: {item['name']}, Qty: {item['quantity']}, Price: R{item['price']}")
    else:
        print("No items found in this category.")

# Inventory summary report
def inventory_summary():
    if not inventory:
        print("The inventory is empty.")
        return
    print("Inventory Summary Report")
    for item in inventory:
        print(f"ID: {item['id']}, Name: {item['name']}, Category: {item['category']}, Qty: {item['quantity']}, Price: R{item['price']}")

# Category summary report
def category_summary():
    if not inventory:
        print("The inventory is empty.")
        return
    category_data = {}
    for item in inventory:
        category = item["category"]
        total_value = item["quantity"] * item["price"]
        if category in category_data:
            category_data[category]["count"] += 1
            category_data[category]["total_value"] += total_value
        else:
            category_data[category] = {"count": 1, "total_value": total_value}

    print("Category Summary Report")
    for category, data in category_data.items():
        print(f"Category: {category}, Number of Products: {data['count']}, Total Value: R{data['total_value']:.2f}")

# Main program loop
while True:
    show_menu()
    choice = input("Enter your choice (1-10): ")

    if choice == "1":
        add_item()
    elif choice == "2":
        view_inventory()
    elif choice == "3":
        search_item()
    elif choice == "4":
        update_item()
    elif choice == "5":
        delete_item()
    elif choice == "6":
        sort_by_price()
    elif choice == "7":
        filter_by_category()
    elif choice == "8":
        inventory_summary()
    elif choice == "9":
        category_summary()
    elif choice == "10":
        print("Exiting program. Goodbye!")
        break
    else:
        print("Invalid choice. Please enter a number between 1 and 10.")
```
